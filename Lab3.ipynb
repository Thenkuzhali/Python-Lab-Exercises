### 1. Create a function prime() that receives an integer and return whether n is prime or not. Print all prime numbers from 1 to 100 by calling prime() function. For example,
#### Prime(1)
#### Prime(2)
#### ....
#### Prime(100)

```python
def prime(num):
    if num>1:
        for i in range(2, int(num/2)+1):
            if (num%i)==0:
                print(num, "is not a prime number")
                break
            else:
                print(num, "is a prime number")
    else:
        print(num, "is not a prime number")
for i in range(1, 101):
    prime(i)
```

### 2. Develop a simple arithmetic calculator for 4 operations. The program should continue calculation until user types 'q' to quit. A sample user interaction can be:
#### Enter operator (q to quit): +
#### Enter value 1: 10
#### Enter value 2: 20
#### Result = 30
### create 4 functions add(), subtract(), multiply() and divide() that receives two values and returns the result of the operation.
### Now, perform the following operations by calling the corresponding functions. Validate your outputs.
#### 1. 10+20
#### 2. 20-5
#### 3. 8* 5
#### 4. 50/30

```python
def add(a, b):
    return a+b
def subtract(a, b):
    return a-b
def multiply(a, b):
    return a*b
def divide(a, b):
    return a/b
```

```python
while True:
    op = input('Enter operator(q to quit):')
    if op.casefold() == 'q':
        break
    else:
        val1 = int(input('Enter Value 1:'))
        val2 = int(input('Enter Value 2:'))
    if op == '+':
        print('Result = ', add(val1, val2))
    elif op == '-':
        print('Result = ', subtract(val1, val2))
    elif op == '*':
        print('Result = ', multiply(val1, val2))
    elif op == '/':
        print('Result = ', divide(val1, val2))
```

```python
print(add(10,20))
print(subtract(20,5))
print(multiply(8,5))
print(divide(50,3))
```

### 3. Create a function factorial() that takes an integer and returns its factorial value.
#### you can create as a non-recursive version of factorial.
#### Also, check factorial of negative number does not exist.
#### Factorial of 0 is 1.
#### Save this Python file as factorial_definition.py.
### Now, open an another file and you can import factorial_definition.py as follows:
#### import factorial_definition
#### you can call factorial function as factorial_definition.factorial().
### Now, print the following values:
#### 1. factorial_definition.factorial(3)
#### 2. factorial_definition.factorial(5)
#### 3. factorial_definition.factorial(10)

```python
def factorial(n):
    fact = 1
    if n<0:
        print("Factorial of negative number does not exists.")
    elif n==0:
        print('Factorial of 0 is 1')
    else:
        while (n>0):
            fact *= n
            n -= 1
        print('Factorial of the number is: ', fact)
```

```python
factorial(0)
```

```python
factorial(3)
```

```python
factorial(5)
```
